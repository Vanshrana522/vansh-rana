```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
df = pd.read_csv('/content/country_wise_latest.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by 'Deaths' and select the top 10 countries
top_10_deaths = df.nlargest(10, 'Deaths')

plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Deaths', data=top_10_deaths, palette='Reds_d', hue='Country/Region', legend=True)
plt.xlabel('Country/Region')
plt.ylabel('Total Deaths')
plt.title('Top 10 Countries by Total COVID-19 Deaths')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you want to plot Deaths vs Recovered from the COVID-19 DataFrame 'df'
sns.lineplot(data=df, x='Deaths', y='Recovered')
plt.title('Total Deaths vs. Recovered Cases')
plt.xlabel('Total Deaths')
plt.ylabel('Total Recovered')
plt.show()
```

```python
sns.relplot(data=df,x = 'Deaths',y = 'Recovered',kind = 'scatter',hue = 'Country/Region')
```

```python
sns.displot(data=df,x = 'Deaths',hue = 'Country/Region',kind = 'hist',bins = 4)
```

```python
sns.relplot(data=df, x = 'Deaths',





            y = 'Recovered', kind = 'line', hue = 'WHO Region')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation analysis
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
