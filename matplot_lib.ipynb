```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#2768F5',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#9F27F5',linestyle='dashdot',linewidth=4)
plt.title('ROHIT SHARMA VS KOHLI CARRER COMPARISION')
plt.xlabel('season')
plt.ylabel('run score')
plt.legend(['V Kohli','RG Sharma'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

batsman = pd.read_csv('/content/sharma-kohli.csv')

plt.figure(figsize=(10, 6))
plt.scatter(batsman['index'], batsman['V Kohli'], color='red', label='V Kohli', marker='X', s=100)
plt.scatter(batsman['index'], batsman['RG Sharma'], color='blue', label='RG Sharma', marker='o', s=100)
plt.title('ROHIT SHARMA VS KOHLI CAREER COMPARISON (Scatter Plot)')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
base_price= pd.read_csv('/content/IPL_2024_Players_Auction_Dataset[1].csv')
# Display the first few rows to check column names
display(base_price.head())

# Plotting 'COST IN ₹ (CR.)' against the DataFrame's index
plt.figure(figsize=(12, 6))
plt.plot(base_price.index, base_price['2023 Squad'], color='#2768F5', linestyle='solid', linewidth=2)
plt.title('IPL 2024 Players Auction: Cost Trend')
plt.xlabel('Player ID (DataFrame Index)')
plt.ylabel('2023 Squad')
plt.grid(True)
plt.show()
```
